```python
# print 0 to 20 by using range

for i in range(0, 20):
    print(i)
```

```python
# print range 10 to 20

for i in range(10, 20):
    print(i, end=" ")
```

```python
# print number of items in the list by using 'len'

my_list = [10, 20, 14, 55, 43, 87, 76]

print(my_list)
print("Number of item in the List:", len(my_list))
```

```python
# Print each character vertically

text = "Artificial Intelligence"
print(text)

for letter in text:
    print(letter)
```

```python
numbers = (20, 10, 16, 19, 25, 1, 276, 188)

for num in numbers:
    if num % 2 != 0:
        print(num, "is odd")
```

```python
numbers = (20, 10, 16, 19, 25, 1, 276, 188)
print(numbers)

for num in numbers:
    if num % 2 != 0:
        print(num, "is odd")
```

```python
numbers = (20, 10, 16, 19, 25, 1, 276, 188)

for num in numbers:
    if num % 2 == 0:
        print(num, "is even")
```
